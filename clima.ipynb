```python
import requests
import pandas as pd
import json
#Pega a data e hora atual do seu computador no momento da execução
from datetime import datetime

# 1 - CONFIGURAR O END-POINT
url = 'https://api.open-meteo.com/v1/forecast'

# Coletar a localização
print("=== COLETA DE DADOS CLIMÁTICOS ===\n")
latitude = float(input("Informe a latitude?\n"))
longitude = float(input("Informe a Longitude?\n"))

# Definir os parâmetros que seram usados
# Temperatura máxima e soma de chuva diária para os últimos 7 dias
parametros = {
    "latitude": latitude,
    "longitude": longitude,
    "current_weather": True,
    "daily": "temperature_2m_max,precipitation_sum",
    "timezone": "auto"
}
print(f"\nLatitude: {latitude}")
print(f"Longitude: {longitude}")
print(f"Enviando requisição para a API...\n")

# 2 - COLETA DE DADOS E VERIFICAÇÃO
try:
    resposta = requests.get(url, params=parametros)

    if resposta.status_code == 200:
        print("Conexão com a API realizada com sucesso!🛸👽🛸\n")

        # Transformando a resposta em JSON
        dados = resposta.json()
        print("Dados foram convertidos para JSON com sucesso!🛸👽🛸\n")

        # Exibir dados atuais
        temperatura_atual = dados['current_weather']['temperature']
        velocidade_vento = dados['current_weather']['windspeed']

        print("=== DADOS ATUAIS ===")
        print(f"Temperatura atual: {temperatura_atual}℃")
        print(f"Velocidade do vento: {velocidade_vento} Km/h\n")

        # 4 - O EXTRA (Fahrenheit)
        temperatura_fahrenheit = (temperatura_atual * 9/5) + 32
        print(f"Temperatura em Fahrenheit: {temperatura_fahrenheit:.2f}°F\n")

        # Navegando no JSON para pegar a parte daily
        dados_diarios = dados['daily']
        datas = dados_diarios['time']
        temperaturas_max = dados_diarios['temperature_2m_max']
        chuvas_diarias = dados_diarios['precipitation_sum']

        print("===🛸 DADOS DIÁRIOS 👽(ÚLTIMOS 7 DIAS)🛸 ===")
        print(f"Período: {datas[0]} até {datas[-1]}\n")

        # Criando o DataFrame
        df = pd.DataFrame({
            'Data': datas,
            'Temperatura_Máxima_Celsius': temperaturas_max,
            'Temperatura_Máxima_Fahrenheit': [(temp * 9/5) + 32 for temp in temperaturas_max],
            'Chuva_Diária_mm': chuvas_diarias
        })

        # Exibir o DataFrame
        print(df.to_string(index=False))
        print("\n")

        # Estatísticas
        print("=== ESTATÍSTICAS 🛸👽🛸===")
        print(f"Temperatura máxima média: {df['Temperatura_Máxima_Celsius'].mean():.2f}℃")
        print(f"Temperatura máxima máxima: {df['Temperatura_Máxima_Celsius'].max():.2f}℃")
        print(f"Temperatura máxima mínima: {df['Temperatura_Máxima_Celsius'].min():.2f}℃")
        print(f"Chuva total: {df['Chuva_Diária_mm'].sum():.2f} mm")
        print(f"Dias com chuva: {(df['Chuva_Diária_mm'] > 0).sum()} dias\n")

        # 3 - SALVAR OS DADOS (O ARQUIVO CSV)
        #Formata a data/hora em texto (string) do jeito que você quiser usando com .strftime
        nome_arquivo = f"clima_lat{latitude}_lon{longitude}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        df.to_csv(nome_arquivo, index=False, encoding='utf-8')
        print(f"✅ Dados salvos com sucesso em: {nome_arquivo}")

    else:
        print(f"Erro ao usar a API")
        print(f"Número do erro: {resposta.status_code}")

except requests.exceptions.RequestException as e:
    print(f" Erro na conexão: {e}")
except KeyError as e:
    print(f"Erro ao acessar dados: {e}")
except Exception as e:
    print(f"Erro inesperado: {e}")
```
